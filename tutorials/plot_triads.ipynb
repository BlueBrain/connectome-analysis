# Connected triad motifs

This notebook generates the image of all connected triads (i.e., subgraphs on three nodes).  It's provided here to clarify the code base we use to count them in a graph. 

In the figure below, we show all the possible isomorphism classes of connected triads.  The color of the nodes is given by the number of edges (from 2 to 6).  We also provide the tuple we use to identify each as well as the  number of graphs in its isomorphism class.

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
label_edges=np.arange(3*3).reshape(3,3)

def _get_triad_id(edges):
    row, col=tuple(zip(*edges))
    return tuple(np.sort(label_edges[row, col]))

#We list all connected triads by hand 
connected_triads={
    #On two edges 
    0:((0,2), (1,0)),
    1:((0,2), (1,2)),
    2:((0,1),(0,2)),
    #On three edges 
    3:((0,1), (1,2), (0,2)),
    4:((0,1), (0,2), (1,0)),
    5:((0,1), (1,0), (2,0)),
    6:((0,1),(1,2), (2,0)),
    #On four edges 
    7:((0,1), (0,2), (1,0), (2,0)),
    8:((0,1), (0,2),(1,0), (2,1)),
    9:((0,1), (0,2),(1,0), (1,2)),
    10:((0,1), (0,2), (1,2), (2,1)),
    #On 5 edge
    11:((0,1), (0,2), (1,0),(1,2), (2,0)),
    #On 6 edges
    12:((0,1), (0,2), (1,0),(1,2), (2,0), (2,1))
}
# Size of isomorphism class of each triad type i.e., number of permutation of the vertices giving the same graph
triad_combinations = np.array([6, 3, 3, 6, 6, 6, 2, 3, 6, 3, 3, 6, 1])
triad_dict ={_get_triad_id(connected_triads[i]):i for i in range(13)}
# The 3-cycle needs to be entered twice because all edges have the same in-out degree but there are two 
# graphs in the isomorphism class
triad_dict[(2, 3, 7)]=6 

#Plotting triads 
pos={0:(0,0), 1:(0,0.5), 2:(1,0)}; node_color="lightgrey"
rows=3;cols=7
fig, axs = plt.subplots(rows,cols, figsize =(15,4.5),squeeze=False, height_ratios=[1,0.2,1])
for k in range(13):
    i=(k//cols)*2; j=k%cols
    G=nx.DiGraph()
    G.add_edges_from(connected_triads[k])
    nx.draw_networkx(G, ax=axs[i,j], arrowstyle='-|>', pos=pos, 
                     node_color=f'C{len(G.edges)-2}', alpha=0.75,width=1.5,
                    arrowsize=20)#, node_shape='o')
    title = f'{_get_triad_id(connected_triads[k])} \n{triad_combinations[k]} isos'
    if k==6: title = f'(1, 5, 6)\nor\n (2,3,7) \n{triad_combinations[k]} isos'
    axs[i,j].set_title(title)

for ax in axs.flatten():
    if ax!= axs[2,6]: ax.axis("off")
axs[2,6].set_title("Edge indices")
#Plotting matrix of edge indices 
ax=axs[2,6]
ax.set_xlim([0, 3])
ax.set_ylim([-2, 1])
ax.set_xticks([0.5,1.5,2.5])
ax.set_yticks([0.5, -.5, -1.5])
for spine in ax.spines.keys(): ax.spines[spine].set_color("lightgrey")
ax.set_xticks([])
ax.set_yticks([])
ax.hlines(y=[0,-1],xmin=0, xmax=3, color="lightgrey", linestyle="--")
ax.vlines(x=[1,2],ymin=-2, ymax=1, color="lightgrey", linestyle="--")
label_edges=np.arange(3*3).reshape(3,3).T
for i in range(3):
    for j in range(3):
        ax.text(i+0.5,-j+0.5, str(label_edges[i,j]), va='center', ha='center', fontsize=12)
fig.suptitle("Connected triad motifs and their id", y=1.1, fontsize=14)
axs[2,2].text(-.2, -.3, "Colors of the nodes represent the number of edges", fontsize=12)
plt.savefig(f"figs/triads.png", dpi=300, transparent=True,bbox_inches='tight')
```
