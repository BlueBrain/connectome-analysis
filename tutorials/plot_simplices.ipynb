```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
#Plotting 2-simplices of all types.
pos={0:(0,0), 1:(0,0.5), 2:(1,0)};
rows=1;cols=3
fig, axs = plt.subplots(rows,cols, figsize =(10,2.5))
#Directed simplex
axs[0].set_title("Directed 2-simplex")
G=nx.DiGraph()
G.add_edges_from(((0,1), (1,2), (0,2)))
nx.draw_networkx(G, ax=axs[0], pos=pos, node_color=f'C0', alpha=0.75,width=1.5,
                arrows=True, arrowsize=20)
#Undirected simplex
axs[1].set_title("Undirected 2-simplex")
G=nx.Graph()
G.add_edges_from(((0,1), (1,2), (0,2)))
nx.draw_networkx(G, ax=axs[1], pos=pos, node_color=f'C1', alpha=0.75,width=1.5)
#Reciprocal simplex
axs[2].set_title("Reciprocal 2-simplex")
G=nx.DiGraph()
G.add_edges_from(((0,1), (1,2), (0,2), (1,0), (2,1), (2,0)))
nx.draw_networkx(G, ax=axs[2], pos=pos, node_color=f'C2', alpha=0.75,width=1.5,
                 arrows=True,arrowsize=20)
for ax in axs.flatten():
    ax.axis("off")
fig.suptitle("Examples of simplex motifs")
plt.tight_layout()
plt.savefig(f"figs/2_simplices.png", dpi=300, transparent=True,bbox_inches='tight')
```
